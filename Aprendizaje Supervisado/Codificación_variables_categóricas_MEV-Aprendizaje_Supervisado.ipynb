Para un modelo base no incluiremos el raster, por lo que sólo codificaremos la variable estación.

```python
enc = OneHotEncoder(handle_unknown='ignore', sparse=False, dtype= int)
```

```python
df_A_copy=df_A.copy()
```

```python
df_B_copy=df_B.copy()
```

```python
df_C_copy=df_C.copy()
```

```python
df_D_copy=df_D.copy()
```

```python
enc.fit(df_A_copy.estacion.values.reshape(-1,1))
enc.fit(df_B_copy.estacion.values.reshape(-1,1))
```

```python

enc.fit(df_C_copy.estacion.values.reshape(-1,1))
enc.fit(df_D_copy.estacion.values.reshape(-1,1))
```

```python
df_A_encoded= enc.transform(df_A_copy.estacion.values.reshape(-1,1))
df_B_encoded= enc.transform(df_B_copy.estacion.values.reshape(-1,1))
```

```python
df_C_encoded= enc.transform(df_C_copy.estacion.values.reshape(-1,1))
df_D_encoded= enc.transform(df_D_copy.estacion.values.reshape(-1,1))
```

```python
encoded_A_columns = enc.get_feature_names(['estacion'])
encoded_B_columns = enc.get_feature_names(['estacion'])
```

```python

encoded_C_columns = enc.get_feature_names(['estacion'])
encoded_D_columns = enc.get_feature_names(['estacion'])
```

```python
df_A_encoder = pd.DataFrame(data=df_A_encoded, columns= encoded_A_columns)
df_B_encoder = pd.DataFrame(data=df_B_encoded, columns= encoded_B_columns)
```

```python

df_C_encoder = pd.DataFrame(data=df_C_encoded, columns= encoded_C_columns)
df_D_encoder = pd.DataFrame(data=df_D_encoded, columns= encoded_D_columns)
```

```python
df_A_xgb=pd.concat([df_A_copy, df_A_encoder], axis=1).drop(['estacion'], axis=1)
```

```python
df_A_xgb.info()
```

```python
df_B_xgb=pd.concat([df_B_copy, df_B_encoder], axis=1).drop(['estacion'], axis=1)
```

```python
df_B_xgb.info()
```

```python
df_C_xgb=pd.concat([df_C_copy, df_C_encoder], axis=1).drop(['estacion'], axis=1)
```

```python
df_D_xgb=pd.concat([df_D_copy, df_D_encoder], axis=1).drop(['estacion'],axis=1)
```
